# Evaluating ML model using K-Fold CrossValidation

- Dataset - Boston House price
- Model - SVR model of SVM
- Cross Validation tech - K-Fold

### 1. Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
```

### 2. Load Data

```python
dataset = pd.read_csv("data//housing.csv", delimiter=r"\s+")
dataset.head()
```

```python
print(dataset.shape)
```

### 3. Preprocessing

```python
X = dataset.iloc[:, [0,12]]
y = dataset.iloc[:, 13]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(X)
```

MimMaxScaler weill re-scale the data between specified range . Here 0-1 to ensure that certain features do not affect the final prediction more than the other features

### 4. KFold CV

```python
scores = []
best_svr = SVR(kernel='rbf')
cv = KFold(n_splits = 10)

for train_index, test_index in cv.split(X):
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)
    
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    best_svr.fit(X_train, y_train)
    scores.append(best_svr.score(X_test, y_test))
```

```python
print(np.mean(scores))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

print("Cross validation score: " ,cross_val_score(best_svr, X, y, cv=10))
```

```python
print("Cross validation prediction: ", cross_val_predict(best_svr, X, y, cv=10))
```
